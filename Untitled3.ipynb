# Data Analisis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("data/ram-test.csv")
df2 = pd.read_csv("data/ram-test2.csv")
```

```python
df = pd.concat([df1, df2], ignore_index=True)
df
```

```python
df.drop('Unnamed: 0', inplace=True, axis=1)
df
```

```python
df.info()
```

```python
df.describe().transpose()
```

```python
sns.set(rc={'figure.figsize': (15, 8)})

boxplot_color = 'green'

sns.boxplot(data=df[['loudness']], color=boxplot_color).set(title='loudness boxplots')

plt.xlabel('Features')
plt.ylabel('Value')

plt.show()
```

```python
sns.set(rc={'figure.figsize': (15, 8)})

boxplot_color = 'orange'

sns.boxplot(data=df[['key']], color=boxplot_color).set(title='key boxplots')

plt.xlabel('Features')
plt.ylabel('Value')

plt.show()
```

```python
sns.set(rc={'figure.figsize': (15, 8)})

boxplot_color = 'blue'

sns.boxplot(data=df[['tempo']], color=boxplot_color).set(title='tempo boxplots')

plt.xlabel('Features')
plt.ylabel('Value')

plt.show()
```

```python
sns.set(rc={'figure.figsize': (15, 8)})

boxplot_color = 'yellow'

sns.boxplot(data=df[['duration']], color=boxplot_color).set(title='duration boxplots')

plt.xlabel('Features')
plt.ylabel('Value')

plt.show()
```

```python
corr_df = df[['loudness', 'num_samples',
       'tempo', 'tempo_confidence', 'key', 'key_confidence', 'duration', 'mode',
       'mode_confidence']].corr()

plt.figure(figsize = (15,8))
heatmap = sns.heatmap(corr_df,
                     annot = True,
                      vmin= -1,
                      vmax = 1,
                     cmap = 'winter',
                     linecolor='Black')
heatmap.set_title('Correlation Heatmap between features')
```

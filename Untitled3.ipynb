# Data Analisis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("data/ram-test.csv")
df2 = pd.read_csv("data/ram-test2.csv")
```

```python
df = pd.concat([df1, df2], ignore_index=True)
df
```

```python
df.drop('Unnamed: 0', inplace=True, axis=1)
df
```

```python
df.info()
```

```python
df.describe().transpose()
```

```python
sns.set(rc={'figure.figsize': (15, 8)})

boxplot_color = 'green'

sns.boxplot(data=df[['loudness']], color=boxplot_color).set(title='Feature boxplots')

plt.xlabel('Features')
plt.ylabel('Value')

plt.show()
```

```python
sns.set(rc={'figure.figsize': (15, 8)})

boxplot_color = 'orange'

sns.boxplot(data=df[['key']], color=boxplot_color).set(title='Feature boxplots')

plt.xlabel('Features')
plt.ylabel('Value')

plt.show()
```

```python
def spotify_token ():
    
    """This function refreshes a 'queries token' for a given app on Spotify.
    queries token are refreshed every 3600 seconds
    returns: token as a string
    """

    # 1. Defining: credentials fot the app
    
    CLIENT_ID = os.getenv("CLIENT_ID")
    CLIENT_SECRET = os.getenv("CLIENT_SECRET")
    
    #2. Request
    
    body_params = {"grant_type":"client_credentials"}
    url = "https://accounts.spotify.com/api/token"
    response = requests.post(url, data=body_params, auth=(CLIENT_ID,CLIENT_SECRET))
    
    try:
        token = response.json()["access_token"]
        return token

    except:
        print("Nope. Maybe check credentials?")
```

```python
def extract_id_from_link(link, website):
    if link.startswith(website):
        start_index = link.find(website) + len(website)
        end_index = link.find("?")

        # Return the extracted ID
        if end_index != -1:
            return link[start_index:end_index]
        else:
            return link[start_index:]
```

```python
def extract_ids_from_links(links, website):
    extracted_ids = []
    for link in links:
        extracted_id = extract_id_from_link(link, website)
        extracted_ids.append(extracted_id)
    return extracted_ids
```

```python
def request_from_spotify(playlist_ids, token):
    headers = {"Authorization": f"Bearer {token}"}
    url_base = "https://api.spotify.com/v1/playlists/"
    responses = []
    for playlist_id in playlist_ids:
        url = url_base + playlist_id
        response = requests.get(url, headers=headers)
        responses.append(response)
    return responses
```

```python
def get_song_ids_from_playlists(playlist_ids, token):
    headers = {"Authorization": f"Bearer {token}"}
    url_base = "https://api.spotify.com/v1/playlists/"
    responses = []
    song_ids = []

    for playlist_id in playlist_ids:
        url = url_base + playlist_id
        response = requests.get(url, headers=headers)
        responses.append(response)

    for response in responses:
        json_data = response.json()

        if "tracks" in json_data and "items" in json_data["tracks"]:
            track_items = json_data["tracks"]["items"]
            for item in track_items:
                if (
                    "track" in item
                    and isinstance(item["track"], dict)
                    and "id" in item["track"]
                ):
                    track_id = item["track"]["id"]
                    song_ids.append(track_id)

    return song_ids
```

```python
song_ids = get_song_ids_from_playlists(music_playlist_all_ids, token)
song_ids
```

```python
def remove_duplicates(lst):
    return list(set(lst))

song_ids_clean = remove_duplicates(song_ids)
len(song_ids_clean)
```

```python
def divide_list(items):
    divided_lists = []
    sublist = []
    
    for item in items:
        sublist.append(item)
        
        if len(sublist) == 500:
            divided_lists.append(sublist)
            sublist = []
    
    if sublist:
        divided_lists.append(sublist)
    
    return divided_lists
```
